## Libraries, Constants and Functions

```python
#Importing the required Libraries, declaring the relevant constants
# and defining the useful functions.
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.integrate import solve_ivp
a=4
b=2
c=1.5
d=3
e1=0.1
e2=0.01
e3=0.001
e4=0.0001
alpha=d/a
delta=0.2

def no_fishing(t,s):
    N=s[0]
    P=s[1]
    A=np.array([[a,(-b)*N],[c*P,-d]])
    return np.dot(A,s)

def with_fishing(t,s):
    N=s[0]
    P=s[1]
    A=np.array([[a-delta,(-b)*N],[c*P,-d-delta]])
    return np.dot(A,s)

def no_fishing_lin(t,s):
    N=s[0]
    P=s[1]
    A=np.array([[0,-1],[alpha,0]])
    return np.dot(A,s)

def fishing_lin(t,s):
    N=s[0]
    P=s[1]
    A=np.array([[0,-1-(delta/d)],[alpha*(1-(delta/a)),0]])
    return np.dot(A,s)

def selective_fishing(t,s):
    N=s[0]
    P=s[1]
    A=np.array([[a-delta,(-b)*N],[c*P,-d]])
    return np.dot(A,s)

def selective_fishing_lin(t,s):
    N=s[0]
    P=s[1]
    A=np.array([[0,-1],[alpha*(1-(delta/a)),0]])
    return np.dot(A,s)

def log_growth(t,s,e):
    N=s[0]
    P=s[1]
    A=np.array([[a-(e*a*N),(-b)*(N)],[c*(P),-d]])
    return np.dot(A,s)

def log_growth_lin(t,s,e):
    N=s[0]
    P=s[1]
    A=np.array([[-e*a*d/c,(-b)*(d/c)],[(a/b)*(c-e*d),0]])
    return np.dot(A,s)
```

## Question 1: Lotka-Volterra System (without fishing)

```python
## Question 1
#This cell of code represents the non linearized model without fishing.
#It has two visualizations: 
#1)Trajectories of Predators and Preys
#2)Phase portrait

t_eval=np.arange(0,100.01,0.01)
Q1=solve_ivp(no_fishing,[0,100],[4,4],t_eval=t_eval)

#Plotting the trajectories
plt.figure(figsize=(12,8))
plt.plot(Q1.t,Q1.y.T[:,1],label="Predators")
plt.plot(Q1.t,Q1.y.T[:,0],label="Prey")
plt.xlabel('t')
plt.ylabel("Number of predators/prey")
plt.title("Trajectories (non linearized model without fishing)")
plt.xlim([0,6])
plt.ylim([0,6])
plt.legend()
plt.show()

#plotting the phase portrait
plt.figure(figsize=(12,8))
plt.plot(Q1.y.T[:,0],Q1.y.T[:,1])
plt.xlabel('N(t)')
plt.ylabel("P(t)")
plt.title("Phase Portrait (non linearized model without fishing)")
plt.show()
```

## Question 8

```python
##Question 8

#defining the Lorenz attractor function based on the equations given in the question
def lorenz(x, y, z, s=10, r=28, b=8/3): 
    """
    Given:
       x, y, z: a point of interest in three dimensional space
       s, r, b: parameters defining the lorenz attractor
    Returns:
       x_dot, y_dot, z_dot: values of the lorenz attractor's partial
           derivatives at the point x, y, z
    """
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return x_dot, y_dot, z_dot


dt = 0.01 #one iteration 
num_steps = 10000 #number of steps taken

# One more step for the initialization of values
xs = np.empty(num_steps + 1)
ys = np.empty(num_steps + 1)
zs = np.empty(num_steps + 1)

# Seting the initial values
xs[0], ys[0], zs[0] = (0., 1., 1.05)

# Step through "time", calculating the partial derivatives at the current point
# and using them to estimate the next point
for i in range(num_steps):
    x_dot, y_dot, z_dot = lorenz(xs[i], ys[i], zs[i])
    xs[i + 1] = xs[i] + (x_dot * dt)
    ys[i + 1] = ys[i] + (y_dot * dt)
    zs[i + 1] = zs[i] + (z_dot * dt)


# Plot
ax = plt.figure(figsize=(12,8)).add_subplot(projection='3d')

ax.plot(xs, ys, zs, lw=0.5)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor")

plt.show()
```
